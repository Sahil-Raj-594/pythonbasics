```python
#14.Write a program to reverse a string using a while loop.
s=input("Enter a string:")
rev=""         #taking empty string to store the reverse value
i=len(s)-1     # indexing from last element

while i>=0:
  rev+=s[i]    #adding last index element to rev string
  i-=1         #fetching another element from backward
print("Reversed string is",rev)

```
